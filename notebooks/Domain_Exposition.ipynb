# Domain Exposition


## Terminology

| Term | Definition |
|-----|------------|
| Loan Default | Failure of a borrower to repay a loan on time. |
| Credit History | A record of a borrower’s past borrowing and repayment behavior used to evaluate trustworthyness when it comes to giving them more credit |
| Limited Credit History | A situation where a borrower has little or no previous credit activity, making traditional credit evaluation difficult. |
| Debt-to-Income Ratio (DTI) | A financial metric comparing a borrower’s total debt payments to their income, often used to evaluate ability to repay loans. |
| Interest Rate | The percentage charged by a lender for borrowing money, typically expressed annually. |
| Loan Term | The length of time over which a borrower must repay a loan. |
| Credit Score | A numerical representation of a borrower’s creditworthiness based on credit history and financial behavior. |
| Default Probability | The predicted likelihood that a borrower will fail to repay a loan. |
| Borrower | The individual or entity receiving a loan from a lender. |
| Loan Attributes | Characteristics of a loan such as loan amount, interest rate, and repayment term. |

## Domain Explanation

This project exists within the domain of credit risk analysis and financial lending. Financial institutions, such as banks and lending platforms, must evaluate whether they should lend money to an individual while minimizing the risk of financial loss. Credit risk analysis is typically performed using a borrower’s credit history, as individuals with strong credit histories are generally less likely to default on loans. However, this approach makes it difficult to evaluate borrowers with limited credit history because they lack the historical data typically used to assess risk. Within this domain, data science techniques are increasingly used to analyze a wide range of financial and behavioral indicators in order to better estimate borrower risk and support more informed lending decisions.
