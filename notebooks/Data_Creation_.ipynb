# Data Creation

## Provenance

The dataset used in this project is the LendingClub loan dataset, obtained from a publicly available repository of peer-to-peer lending records. Two datasets were provided: accepted_2007_to_2018.csv.gz and rejected_2007_to_2018.csv.gz. This project focuses only on the accepted loans dataset, as the objective is to predict loan default risk, which can only be observed for borrowers who were actually issued loans. The data was downloaded in compressed CSV format (.csv.gz) to efficiently store and manage the large volume of records.

To process the dataset efficiently without exceeding memory limitations, the data was loaded using DuckDB, which allows for querying directly from compressed files. During ingestion, irregularities in the raw data, such as malformed rows, were handled by ignoring parsing errors. Additionally, all fields were initially treated as string data types to prevent type conversion issues. After loading, a subset of relevant features related to borrower financial characteristics, credit history, and loan attributes was selected to create a cleaned dataset suitable for analysis and predictive modeling.

## Bias Identification


Firstly, this data set only includes people who applied for loans through lending club.  This is a form of selection bias as this is not representative of the entire population.  Likely, there may be certain correlations between those who use lending club as compared to those who do not.  

As well, the data being used is those who were already approved for a loan.  This likely means the majority of applicants had fairly good risk scores as otherwise they would not have been given a loan.  

There may also be bias in how people report certain variables.  For example, some people may classify their income or employment information differently than another peorson would.

Finally, as my goal is to focus on those who have low credit history, there may be bias as the subgroup is not representative of the whole data set meaning some patterns may be exagerated if they are specific to that population.

## Bias Mitigation


There are a couple of ways to mitigate some of the bias in the data set.  First of all, providing proper disclaimers is critical.  For example, agknowledging that the results are interpreted as conditional on approved loans rather than generalized to all potential borrowers.

To handle the sampling bias, there can be some analysis looking at the subgroup compared to the broader data set to understand how the distributions differ.  As well, when calculating model perfomrance  it will be important to ensure it is evaluated carefully as to ensure that it is not misleading due to class imbalance or restricted sample size.

Lastly, when testing the model it will be important to see not only how the model performs as a whole but also how it performs on subgroups to ensure that it is not disproportionatly labeling subgroups as risky or unrisky.  

## Rationale


One of the first importat decisions was choosing to just use the accepted loans dataset from lendingclub since loan outcomes are the only way to see if someone will actually default on their loan.  While this introduces bias as this only includes prescreened applicants, ultamelty it is the only way to determine if someone will actually end up defaulting.

Second, as the data set was big and inially compressed, duckdb was chosen to ensure that it was able to be efficiently queried.  This could cause issues in things like parsing errors, but ultametly allowed for scalability and efficiency.

Treating all variables initially as strings during ingestion was a deliberate choice to prevent data type errors. However, this requires later type conversion, which introduces the possibility of misclassification or loss of precision if not handled carefully.

One of the biggest key decisions was selecting a subset of relevent features from the full dataset which has over 100 variables.  The features chosen focused on borrowers financial characteristics (like annual income), credit history indicators (ex fico score), and loan attributes.  This choice was made to reduce noise and increase efficiency while still maintaining enough variables to create a good model.  

Several variables were intentionally excluded.  For example, zip code was excluded as to not create systematic bias.  As well, variables that may not be known at the time of loan approval were excluded like last time a payment was made on the loan.  As well, text heavy or unstructured fields were ommited as they would require a lot of additional processing.  While title of the loan may be relevent, it ultametly would require more computing power to interpret than is beneficial at this time.
