```python
import matplotlib.pyplot as plt

# Data
categories = ["No Credit History", "Unscorable Credit History"]
values = [26, 19]

# Create chart
plt.figure(figsize=(6,4))
plt.bar(categories, values)
plt.ylabel("Millions of Americans")
plt.title("Americans Without a Credit Score")

# Add labels on bars (optional but nice)
for i, v in enumerate(values):
    plt.text(i, v + 0.5, str(v), ha='center')

# Save file (IMPORTANT)
plt.savefig("credit_invisible_chart.png")

plt.show()
```
